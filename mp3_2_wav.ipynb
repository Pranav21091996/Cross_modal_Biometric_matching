```python
import os
import librosa
from pydub import AudioSegment

mp3_path = '/home/pranav/Desktop/BTechProject/mp3'
wav_path = '/home/pranav/Desktop/BTechProject/wav'


for folder in os.listdir(mp3_path):

    os.mkdir(os.path.join(wav_path,folder))

    for file in os.listdir(os.path.join(mp3_path,folder)):
        new_name = file.split('.')
        os.rename(os.path.join(os.path.join(mp3_path,folder),file),os.path.join(os.path.join(mp3_path,folder),new_name[0]+'.mp3'))
        print(file)

        sound = AudioSegment.from_file(os.path.join(os.path.join(mp3_path,folder),file))
        sound.export(os.path.join(os.path.join(mp3_path,folder),file), format="mp3")

        sound = AudioSegment.from_mp3(os.path.join(os.path.join(mp3_path,folder),file))
        sound.export(os.path.join(os.path.join(wav_path,folder),new_name[0]+'.wav'), format="wav")
```
